# load packages

```python
import pandas as pd
```

# read in input files

```python
afr = pd.read_csv('output/AFR.INV_NORMAL_TSH.5_inputs.no_correction.union.metasoft_output.no_mean_hetero_correction.cleaned.txt',
                  sep = '\t')
afr.head()
```

```python
eur = pd.read_csv('output/EUR.INV_NORMAL_TSH.5_inputs.no_correction.union.metasoft_output.no_mean_hetero_correction.cleaned.txt',
                  sep = '\t')
eur.head()
```

```python
afr_vep = pd.read_csv('vep_output/AFR.INV_NORMAL_TSH.5_inputs.no_correction.union.metasoft_output.no_mean_hetero_correction.vep_output.cleaned.txt',
                      sep = '\t')
afr_vep.head()
```

```python
eur_vep = pd.read_csv('vep_output/EUR.INV_NORMAL_TSH.5_inputs.no_correction.union.metasoft_output.no_mean_hetero_correction.vep_output.cleaned.txt',
                      sep = '\t')
eur_vep.head()
```

# clean files
- remove variants in 1 study
- extract ID, CHR, BP, REF, ALT, pval, and beta columns
- add RSID and gene from VEP output

```python
afr_sub = afr[afr['#STUDY'] > 1]
afr_sub = afr_sub[['RSID', 'PVALUE_RE', 'BETA_RE', 'STD_RE', 'CHR', 'BP', 'REF', 'ALT']]
afr_sub = afr_sub.rename(columns = {'RSID' : 'ID', 'BP' : 'POS'})
afr_vep_sub = afr_vep[['RSID', 'GENE', 'ID']]
afr_merge = afr_vep_sub.merge(afr_sub, on = 'ID', how = 'right')
afr_merge['RSID'] = afr_merge['RSID'].fillna(afr_merge['ID'])
afr_merge = afr_merge.drop(columns = ['ID'])
afr_merge = afr_merge.sort_values(by = ['CHR', 'POS'])
print(len(afr.index))
print(len(afr_sub.index))
print(len(afr_merge.index))
afr_merge.head()
```

```python
eur_sub = eur[eur['#STUDY'] > 1]
eur_sub = eur_sub[['RSID', 'PVALUE_RE', 'BETA_RE', 'STD_RE', 'CHR', 'BP', 'REF', 'ALT']]
eur_sub = eur_sub.rename(columns = {'RSID' : 'ID', 'BP' : 'POS'})
eur_vep_sub = eur_vep[['RSID', 'GENE', 'ID']]
eur_merge = eur_vep_sub.merge(eur_sub, on = 'ID', how = 'right')
eur_merge['RSID'] = eur_merge['RSID'].fillna(eur_merge['ID'])
eur_merge = eur_merge.drop(columns = ['ID'])
eur_merge = eur_merge.sort_values(by = ['CHR', 'POS'])
print(len(eur.index))
print(len(eur_sub.index))
print(len(eur_merge.index))
eur_merge.head()
```

# export

```python
afr_merge.to_csv('locus_zoom/AFR.INV_NORMAL_TSH.5_inputs.metasoft.locus_zoom_input.txt',
                 sep = '\t',
                 index = None,
                 na_rep = 'NaN')
```

```python
eur_merge.to_csv('locus_zoom/EUR.INV_NORMAL_TSH.5_inputs.metasoft.locus_zoom_input.txt',
                 sep = '\t',
                 index = None,
                 na_rep = 'NaN')
```
